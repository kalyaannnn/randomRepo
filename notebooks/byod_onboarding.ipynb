# BYOD Onboarding Demo

This notebook shows the canonical v1 path for bringing your own dataset, starting from `BYODRecord` and `make_single_turn_task(...)`.


Use the high-level API first for single-turn tasks. Drop to `BaseEnvironment` and `BaseVerifier` when you need custom multi-turn state transitions or another lower-level contract.


```python
from agentrl import BYODRecord, GRPOConfig, GRPOTrainer, SFTBootstrapTrainer, make_single_turn_task

records = [
    BYODRecord(
        input="Return exactly: ok",
        reference_answer="ok",
        supervised_target="ok",
        metadata={"split": "train"},
    )
]

task = make_single_turn_task(
    records=records,
    prompt_formatter=lambda record, tokenizer: f"User:\n{record.input}\n\nAssistant:\n",
    reward_fn=lambda response, state: 1.0 if response.strip() == state["reference_answer"] else 0.0,
    supervised_target_fn=lambda record: record.supervised_target,
)

config = GRPOConfig(
    model_name="Qwen/Qwen2.5-0.5B-Instruct",
    steps=2,
    batch_size=1,
    group_size=2,
    max_new_tokens=32,
)
```

The notebook intentionally keeps the training object construction visible without wiring up a full run. That keeps the onboarding path focused on the API shape.
